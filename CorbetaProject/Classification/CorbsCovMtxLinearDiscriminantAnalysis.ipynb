```python
#Covariance Matrix Linear Discriminant Analysis
# Generate some synthetic data
X, y = make_classification(n_samples=100, n_features=4, n_classes=2, random_state=42)

# Compute the class means
class_means = np.mean(X, axis=0)

# Compute the within-class scatter matrix
S_w = np.zeros((X.shape[1], X.shape[1]))
for label in np.unique(y):
    X_c = X[y == label]
    class_mean = np.mean(X_c, axis=0)
    centered_data = X_c - class_mean
    scatter_matrix = np.dot(centered_data.T, centered_data)
    S_w += scatter_matrix

# Compute the between-class scatter matrix
S_b = np.zeros((X.shape[1], X.shape[1]))
for label in np.unique(y):
    n = X[y == label].shape[0]
    class_mean_diff = class_means - np.mean(X[y == label], axis=0)
    S_b += n * np.outer(class_mean_diff, class_mean_diff)

# Compute the eigenvectors and eigenvalues of S_w^(-1) S_b
eigenvalues, eigenvectors = np.linalg.eig(np.dot(np.linalg.inv(S_w), S_b))

# Sort the eigenvectors by eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Choose the number of dimensions for the LDA projection
n_dimensions = 1

# Construct the projection matrix W
W = eigenvectors[:, :n_dimensions]

# Project the data onto the LDA subspace
X_lda = np.dot(X, W)

# Print the projected data
print(X_lda)
```

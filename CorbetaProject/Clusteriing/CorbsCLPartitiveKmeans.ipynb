```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the CSV file using pandas
data = pd.read_csv('F:\\CS 325 Data Mining\\Datasets\\archive\\Heart.csv')
data.head()
```

```python
X= data.iloc[:,1:14]
Y= data.iloc[:,-1]
data['ChestPain'].replace(['typical', 'nontypical', 'nonanginal', 'asymptomatic'], [1,2,3,4], inplace=True)
data['Thal'].replace(['normal', 'fixed', 'reversable'], [1,2,3], inplace=True)
data['AHD'].replace(['No', 'Yes'], [0,1], inplace=True)
data.replace('NA', np.nan, inplace = True)
data.replace(' ', np.nan, inplace = True)
data.dropna( inplace = True, how = 'any')
data = data[data.Thal != '.']
data.head()
```

```python
# Separate the features (X)
X = data.iloc[:, :-1]  # Assuming the target variable is not included in the last column
```

```python
# Scale the input features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Perform partition-based k-means clustering
k = 3  # Specify the number of clusters
kmeans = KMeans(n_clusters=k)  
labels = kmeans.fit_predict(X_scaled)
```

```python
# Add the cluster labels to the original data
data['Cluster'] = labels
```

```python
# Calculate the distances, nearest cluster, and new centroid
distances = kmeans.transform(X_scaled)
nearest_cluster = np.argmin(distances, axis=1)
new_centroid = kmeans.cluster_centers_[nearest_cluster]
```

```python
# Display the information
data['Distance 1'] = distances[:, 0]
data['Distance 2'] = distances[:, 1]
data['Nearest Cluster'] = nearest_cluster
data['New Centroid'] = list(new_centroid)

# Print the updated data with the additional information
print(data)
```
